2. Solve the following system of linear equations using Gaussian elimination (Ax=y) 
x + 2y + 3x = 9 
2x - y + 2z = −3 
3x + 3y + z = 5 


```r
library(matlib)
library(pracma)
A <- matrix(c(1, 2, 3,
            2, 1, 2,
            3,  3, 1), 
            3, 3, byrow=TRUE)
b <- c(9, -3, 5)
result <- gaussianElimination(A, b)
result
paste("x =", result[10], ' y = ', result[11], ' z =', result[12])
```

4. Get the COVID-19 Dataset from the data sources. Number of observations should be more than 100.Then, report the following informations 
a. Data Source detail(Ex: Link) 
b. Explain the Unit & Necessity of each variable 
c. Find the missing values(rows & columns) and replace them with mean(Tidy Dataset) 
d. Generate the two new variables(Var1:Mean, Var2: Median from available variable) 
e. Rename the two existing variables 
f. Create a plot using following instructions (using 7 layers of Grammar of Graphics) 
    i. Choose x and y axis(aes) 
    ii. geom_point() - specify the parameters, size : 5, color: red, alpha: ⅕ 
    iii. Use Facet grid, cartesian coordinates & geom_smooth() 
    iv. Assign the title to x, y and graph 
    v. Export the graph to your working directory with the title called “covid_19_dataset.png”


```r
# a. Data Source detail (https://data.europa.eu/euodp/en/data/dataset/covid-19-coronavirus-data)
# b. Unit and Necessity of each variable
# dateRep - The full date reported
# day - The day
# month - The month
# year - The year
# cases - The number of positive cases
# deaths - The number of deaths
# countriesAndTerritories - the names of the countries and territories
# geoId - country acronym
# countryterritoryCode - country code
# popData2019 - The total population recorded in 2019
# continentExp - The continent that the country is on
# Cumulative_number_for_14_days_of_COVID-19_cases_per_100000 - total amount of case per 100000 people over 14 days
library(tidyverse)
dataset <- read_csv("../Covid-19_Data_Worldwide.csv")
Data <- dataset
Data
names(Data)
```

```r
# c. Find the missing values(rows & columns) and replace them with mean(Tidy Dataset) 
Data[] <- lapply(Data, function(x) { 
  x[is.na(x)] <- mean(x, na.rm = TRUE)
  x
})
Data
```

```r
# d. Generate the two new variables(Var1:Mean, Var2: Median from available variable) 
Data %>% mutate(Mean = mean(cases), Median = median(cases)) 
```

```r
# e. Rename the two existing variables 
Data %>% rename(Population2019 = popData2019, Continent = continentExp)
```

```r
# f. Create a plot using following instructions (using 7 layers of Grammar of Graphics)
# i. Choose x and y axis(aes) - x as Reported Date, y as Positive Cases
data <- dataset %>% group_by(dateRep) %>% count(dateRep, wt = cases)
Date <- data[['dateRep']]
Positive <- data[['n']]
data
```

```r
# ii. geom_point() - specify the parameters, size : 5, color: red, alpha: ⅕ 
library(ggplot2)
gp = geom_point(alpha=1/5, color="red", size=5)
gp
p <- ggplot(x = data, mapping = aes(x = Date, y = Positive)) + gp
p
```

```r
# iii. Use Facet grid, cartesian coordinates & geom_smooth() 

# facet_grid
p + facet_grid(Positive)

# cartesian coordinates
p + coord_cartesian(expand = FALSE)

# geom_smooth
p + geom_smooth()
```

```r
# iv. Assign the title to x, y and graph 
p <- p + ggtitle("Number of Positive Covid-19 case Worldwide by different Countries and Dates") +
  xlab("Date") + ylab("Number Of Positive Cases")
p
```

```r
# v. Export the graph to your working directory with the title called “covid_19_ dataset.png”
ggsave(filename = "./covid_19_dataset.png", units = "cm", width = 25, height = 25)
```
