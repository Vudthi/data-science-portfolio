```r
# Tidy dataset
data <- ggplot2::mpg
help("mpg")
```

```r
# Create new column
library(dplyr)
data$Transmission <- ifelse(startsWith(data$trans,'a'), "automatic", "manual")
data
```

```r
# Plot the graph and customization
library(ggplot2)
graph <- ggplot(data, aes(x=cty, y=hwy))+xlab("City Miles Per Gallon")+ylab("Highway Miles Per Gallon")+ggtitle("Fuel Economy Data from 1999 to 2008")+facet_grid(Transmission ~ year)+theme(axis.title.x=element_text(color="black", size=15, face="bold", margin=margin(t=25)),
      axis.title.y=element_text(color="black", size=15, face="bold", margin=margin(r=30)),
      plot.title=element_text(size=20, color="black", face="bold", hjust=0.5, margin=margin(b=20)),
      strip.background=element_rect(color="red", fill="yellow"))+geom_col(fill="yellow", color="blue")
graph
```
